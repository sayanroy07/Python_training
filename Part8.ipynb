```python
# Group By
df.groupby(by=['Region'])['Sales'].mean().sort_values(ascending=False).plot(kind='barh')
```

```python
# Group By
df.groupby(by=['Product Category','Product Sub-Category'])['Profit'].mean().sort_values(ascending=False).plot(kind='barh')
```

```python
# Group By - Aggregate
df.groupby(by=['Product Category']).aggregate({'Profit':'sum','Sales':'mean','Discount':'max'})
```
