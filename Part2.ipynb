### 4 types of Function Arguments
 - Default Arguments
 - Keyword Arguments
 - Arbitrary Arguments
 - Required/Positional Arguments

```python
# Default Arguments
# A default argument is a parameter that assumes a default value
# if a value is not provided in the function call for that argument.
def welcome(name,greet="Hi"):
    print(greet,name)

welcome("Ranjith")
welcome("Ranjith","Hello")
```

```python
# Keyword Arguments
# The idea is to allow the caller to specify the argument name with values 
# so that the caller does not need to remember the order of parameters.
def welcome(name, greet):
    print(greet, name)


# Keyword arguments
welcome(name = 'John', greet = 'Hello')
welcome(name='Practice', greet='Geeks')
```

```python
# Positional/Required Arguments
# The idea is to have all arguments as per the need
# in a particular positional order
def welcome(name, greet):
    print(greet, name)


# Positional arguments
#welcome('John','Hello')
welcome('John')
#welcome('Practice','Geeks')
```

```python
# Arbitrary Arguments
# In Python it enable functions to 
# accept an unlimited number of arguments
def welcome(*name):
    for i in range(len(name)):
        print(name[i])
    print('\n')


# Positional arguments
welcome('John','Hello')
welcome('John','Hello','Practice','Geeks')
```

```python
# Lambda
# One line anonymous function, which doesn't need to be defined
# Lambda Argument: Expression
mult = lambda x,y: x*y
mult(15,10)
```

```python
sum = lambda x,y,z: x+y+z
sum(5,6,7)
```
